## Building a Model

```python
# Dropping 'DATE' columns from df1
df1.drop('DATE', axis=1, inplace = True)
df1.head(2)
```

```python
df1.shape
```

```python
x  = df1.drop('HPI', axis=1)
x.head(2)
```

```python
y = df1.loc[:, ['HPI']]
y.head(2)
```

```python
len(y)
```

### Splitting the data

```python
from sklearn.model_selection import  train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

### Applying Linear Regression model

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
lr.score(x_test, y_test)
```

```python
lr_pred = lr.predict(x_test)
lr_pred
```

### Applying Cross Validation - KFold to measure accuracy

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

cv_kf = KFold(n_splits =5, random_state=1, shuffle=True)
cross_val_score(LinearRegression(), x, y=y, cv= cv_kf)
```

_**Acurracy of atleast 85%**_

### Applying Cross Validation - ShuffleSplit to measure accuracy

```python
from sklearn.model_selection import ShuffleSplit

cv_ss = ShuffleSplit(n_splits =5, random_state=1, test_size=0.2)
cross_val_score(LinearRegression(), x, y=y, cv= cv_ss)
```

_**Accuracy of atleast 85%, also here average accuracy seems to be improved compared to KFold CV**_

### Finding the best model through GridSearchCV

```python
from sklearn.model_selection import GridSearchCV

from sklearn import svm
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor


def find_best_model_using_gridsearchcv(x,y):
    algos ={
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                'fit_intercept': [True, False]
                 }
        },
        'svm.svr' : {
            'model': svm.SVR(),
            'params': {
                'C': [0.1,1],
                'kernel': ['rbf', 'poly']
                 }
        },
        'ridge': {
            'model': Ridge(),
            'params': {
                'alpha': [1,2],
                'solver': ['svd', 'cholesky']
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random'],
                'max_features':["log2","sqrt"],     
            }
        },
        
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo, config in algos.items():
        gs = GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(x,y)
        scores.append({
            'model':algo,
            'best_score':gs.best_score_,
            'best_params':gs.best_params_
        })
    
    return pd.DataFrame(scores, columns = ['model', 'best_score', 'best_params'])

find_best_model_using_gridsearchcv(x,y)
```

_**It seems that "Lasso" algorithm with parameters alpha=2, selection=cyclic is giving the best score. We will use the same to predict the HPI**_

### Predicting the HPI

```python
lasso =Lasso(alpha =2, selection = 'cyclic')
lasso.fit(x_train, y_train)
print ("Lasso model:", (lasso.coef_))
```

```python
def predict_HPI(Unemployment_Rate,GDP,House_Permits,Population_Growth,Recession,House_Tax,Total_Construction_Spending,Life_Expectancy,Mortgage_Rate_US,CPI,Wilshire_REIT_Index_US,Housing_Inventory_Estimate):
    z = np.zeros(12)
    z[0] = Unemployment_Rate
    z[1] = GDP
    z[2] = House_Permits
    z[3] = Population_Growth
    z[4] = Recession
    z[5] = House_Tax
    z[6] = Total_Construction_Spending
    z[7] = Life_Expectancy
    z[8] = Mortgage_Rate_US
    z[9] = CPI
    z[10] = Wilshire_REIT_Index_US
    z[11] = Housing_Inventory_Estimate
    
    return lasso.predict([z])[0]
```
